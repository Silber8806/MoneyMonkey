```python
import pandas as pd
import numpy as np
import statsmodels as stats
import matplotlib as plt
import os
import datetime as dt
import requests
import bs4 as bs
import pandas_datareader as pdr
import csv
import yfinance as yf
```

```python
wikipedia_url = 'http://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
r = requests.get(wikipedia_url)

if r.status_code == 200:
    soup = bs.BeautifulSoup(r.text, 'lxml')
    table = soup.find('table', {'class': 'wikitable sortable'})
    tickers = []
    
    for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text
        tickers.append(ticker)

    start = dt.datetime(2010, 1, 1)
    end = dt.datetime.now()
    ticker_frames = []
    if not os.path.exists('stocks'):
        os.mkdir('stocks')
        
    for ticker in tickers:
        ticker = ticker.replace('\n','').replace('-','').replace('.','')
        df = yf.download(ticker, start = start, end = end)
        df.reset_index()
        df['ticker'] = ticker
        ticker_frames.append(df)
        file_name = os.path.join('stocks',ticker + '.csv')
        df.to_csv(file_name)
        
stocks = pd.concat(ticker_frames)
        

        
```

```python
file_name = os.path.join('stocks', 'all.csv')
stocks.to_csv()
```
